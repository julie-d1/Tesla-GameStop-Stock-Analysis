# Tesla-GameStop-Stock-Analysis

```python
import yfinance as yf

# Step 1: Extract Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="max")

# Step 2: Reset the index
tesla_data.reset_index(inplace=True)

# Step 3: Display the first five rows
tesla_data.head()
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Step 1: Download the webpage
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
html_data = requests.get(url).content

# Step 2: Parse the data using BeautifulSoup
soup = BeautifulSoup(html_data, 'html.parser')

# Step 3: Extract the Tesla revenue table and create a DataFrame
tesla_revenue = pd.read_html(str(soup))[0]

# Step 4: Display the last five rows
tesla_revenue.tail()
```

```python
# Step 1: Extract GameStop stock data
gme = yf.Ticker("GME")
gme_data = gme.history(period="max")

# Step 2: Reset the index
gme_data.reset_index(inplace=True)

# Step 3: Display the first five rows
gme_data.head()
```

```python
# Step 1: Download the webpage
url_2 = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"
html_data_2 = requests.get(url_2).content

# Step 2: Parse the data using BeautifulSoup
soup_2 = BeautifulSoup(html_data_2, 'html.parser')

# Step 3: Extract the GameStop revenue table and create a DataFrame
gme_revenue = pd.read_html(str(soup_2))[0]

# Step 4: Display the last five rows
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

# Plotting Tesla's stock closing price
plt.figure(figsize=(10,6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla Stock Price')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.title('Tesla Stock Closing Price Over Time')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# Plotting GameStop's stock closing price
plt.figure(figsize=(10,6))
plt.plot(gme_data['Date'], gme_data['Close'], label='GameStop Stock Price', color='orange')

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Closing Price ($)')
plt.title('GameStop Stock Closing Price Over Time')
plt.grid(True)
plt.legend()
plt.show()
```
